# Task 7
How are the popularity of the movies related to the movie budget?
Find any correlation

```python
data[['title','budget','popularity','vote_average']].sort_values(by = 'budget') #Movies sorted by budget value
```

#### Scatter Plot showing any correlation between Budget and Popularity

```python
import matplotlib.pyplot as plt

#Adding the Data
x = data['budget'].to_numpy()
y = data['popularity'].to_numpy()

#Add an Axes to the current figure and make it the current Axes.
ax = plt.axes()

#creating the scatter plot
ax.scatter(x,y, c = 'y', edgecolors="k")

#adding a title and axis labels
ax.set_title('Popularity of Movies vs Budget')
ax.set_ylabel('Budget (thousands)')
ax.set_xlabel('Popularity Score')

#setting the x and y scale
ax.set_ylim(0,400)

#ax.set_xlim(0.0,4.0)
#why doesnt this x limit work?


#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x, y, 1)

#add linear regression line to scatterplot 
plt.plot(x, m*x+b, c='b')


plt.show() #displays the graph
```

```python
#using seaborn to graph
import seaborn as sns

#scatter plot
sns.scatterplot(x='budget', y='popularity',data = data)
plt.show()

#adding a title and axis labels
ax.set_title('Popularity of Movies vs Budget')
ax.set_ylabel('Budget (thousands)')
ax.set_xlabel('Popularity Score')

#correltion
correlation = np.corrcoef(data['budget'],data['popularity'])[0,1]
print(correlation)

#interpretation
if correlation > 0.5:
    print('There is a srong positive correlation')
elif correlation < -0.5:
    print('There is a srong negative correlation')
else:
    print('Little to no correlation')
```

There seems to be a slight correlation with the budget and popularity of the movie. Although there are many outliers, the linear regression line seems to show some form of correlation. The higher the budget, the higher the popularity score 

#### Scatter Plot showing any correlation between Voting Average and Budget

```python
#Adding the Data
x1 = data['budget'].to_numpy()
y1 = data['vote_average'].to_numpy()

#Add an Axes to the current figure and make it the current Axes.
ax1 = plt.axes()

#creating the scatter plot
ax1.scatter(x1,y1, c = 'c', edgecolors="k")

#adding a title and axis labels
ax1.set_title('Vote Average vs Budget')
ax1.set_ylabel('Budget (thousands)')
ax1.set_xlabel('Vote Average')

#setting the x and y scale
ax1.set_ylim(2,8)

ax.set_xlim(0.0,5.0)
#why doesnt this x limit work?


#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x1, y1, 1)

#add linear regression line to scatterplot 
plt.plot(x1, m*x+b, c='r')


plt.show() #displays the graph
```

```python
#using seaborn to graph
import seaborn as sns

#scatter plot
sns.scatterplot(x='budget', y='buget',data = data)
plt.show()

#correltion
correlation = np.corrcoef(data['budget'],data['vote_average'])[0,1]
print(correlation)

#interpretation
if correlation > 0.5:
    print('There is a srong positive correlation')
elif correlation < -0.5:
    print('There is a srong negative correlation')
else:
    print('Little to no correlation')
```
