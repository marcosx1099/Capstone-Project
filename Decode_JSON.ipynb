# Decoding JSON COLUMN

```python
import numpy as np
import pandas as pd
import json #module needed to decode json format columns
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DS1_C7_Part1_Data_Analysis_Using_Python_Dataset.csv')#importing the csv file

genres = data['genres'] #pulling the 'genres' column

display(genres)
```

```python
column_list = genres.tolist() #converting the genres column to a list
type(column_list)

new_list = [] #creating a new empty list

for y in genres:#for loop that iterates through each value in the list
        
    marcos = pd.read_json(y) #converts each value to json readable format
    new_list.append(marcos.iloc[:,-1:].values.tolist())
data['col'] = new_list #adding this list as a new column in the existing dataframe

display(data)
```

```python
#Adding the Data
x1 = data['revenue'].to_numpy()
y1 = data['popularity'].to_numpy()
#Add an Axes to the current figure and make it the current Axes.
ax1 = plt.axes()

#creating the scatter plot
ax1.scatter(x1,y1, c = 'c', edgecolors="k")

#adding a title and axis labels
ax1.set_title('Popularity vs Revenue')
ax1.set_ylabel('Popularity')
ax1.set_xlabel('Revenue')


#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(x1, y1, 1)

#add linear regression line to scatterplot 
plt.plot(x1, m*x1+b, c='r')


plt.show() #displays the graph
```
